```python
# 1. Import Data
# Import libraries:
import pandas as pd
import numpy as np
```

```python
# Load the data and store it in dataframe df:
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
# 2. Analyzing Individual Feature Patterns Using Visualization
# To install Seaborn we use pip, the Python package manager.
%%capture
!pip install seaborn
```

```python
# Import visualization packages "Matplotlib" and "Seaborn"
# "%matplotlib inline" to plot in a Jupyter notebook
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# list the data types for each column
print(df.dtypes)
```

```python
# Calculate the correlation between variables of type "int64" or "float64" using the method "corr"
corr = df.corr()
corr
```

```python
# Find the scatterplot of "engine-size" and "price"
# Engine size as potential predictor variable of price
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
# As the engine-size goes up, the price goes up: 
# this indicates a positive direct correlation between 
# these two variables. Engine size seems like a pretty good 
# predictor of price since the regression line is almost 
# a perfect diagonal line.

# We can examine the correlation between 'engine-size' and 'price' and 
# see that it's approximately 0.87.
```

```python
df[["engine-size", "price"]].corr()
```

```python
# Find the scatterplot of "highway-mpg" and "price"
sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
# As highway-mpg goes up, the price goes down: 
# this indicates an inverse/negative relationship 
# between these two variables. 
# Highway mpg could potentially be a predictor of price.

# We can examine the correlation between 'highway-mpg' 
# and 'price' and see it's approximately -0.704.
```

```python
df[['highway-mpg', 'price']].corr()
```

```python
# Check if "peak-rpm" is a predictor variable of "price"
sns.regplot(x="peak-rpm", y="price", data=df)
```

```python
# Peak rpm does not seem like a good predictor of 
# the price at all since the regression line is close 
# to horizontal. Also, the data points are very scattered 
# and far from the fitted line, showing lots of variability. 
# Therefore, it's not a reliable variable.

# We can examine the correlation between 'peak-rpm' 
# and 'price' and see it's approximately -0.101616.
```

```python
df[['peak-rpm','price']].corr()
```

```python
# Find the correlation between x="stroke" and y="price"
sns.regplot(x="stroke",y="price", data=df)
df[['stroke','price']].corr()
```

```python
# The relationship between "body-style" and "price"
sns.boxplot(x="body-style", y="price", data=df)
```

```python
# The relationship between "engine-location" and "price"
sns.boxplot(x="engine-location", y="price", data=df)
```

```python
# the relationship between "drive-wheels" and "price"
sns.boxplot(x="drive-wheels", y="price", data=df)
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
